```python
import numpy as np
import pandas as pd

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

from sklearn import metrics
import sklearn.metrics as metrics

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = pd.read_csv('diabetes.csv')
dataset.head()
```

```python
# Normalization
# Assigning Independent and dependent variables
X = dataset.iloc[:, 1:8].values
y = dataset.iloc[:, [8]].values

# Normalization
minmax = preprocessing.MinMaxScaler()
minmax.fit(X)
X_minmax = minmax.transform(X)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_minmax, y, test_size=0.2)
```

```python
model_1=Sequential()
model_1.add(Dense(8, input_dim=len(X[0, :]), activation='sigmoid'))
model_1.add(Dense(4, activation='sigmoid'))
model_1.add(Dense(1, activation='sigmoid'))
```

```python
model_1.summary()
```

```python
model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
model_1.fit(x=X_train, y=y_train, epochs=1000, verbose=1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred_2 = model_1.predict(X_test)
y_pred_2
conconfusion = metrics.confusion_matrix(y_test, y_pred_2.round())
print(f"Confusion matrix:\n{conconfusion}")
```

```python
Accuracy = metrics.accuracy_score(y_test, y_pred_2.round())
print(f'Accuracy: {Accuracy}')
```

```python
Sensitivity_recall = metrics.recall_score(y_test, y_pred_2.round())
print(f'Sensitivity_recall: {Sensitivity_recall}')
```

```python
F1_score = metrics.f1_score(y_test, y_pred_2.round())
print(f'F1_score: {F1_score}')
```

```python
E = metrics.mean_squared_error(y_test, y_pred_2.round())
print(f'E: {E}')
```
